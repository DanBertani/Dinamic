```python
from jupyter_dash import JupyterDash
from dash import dcc, html, Input, Output
import yfinance as yf 
import plotly.express as px
```

```python
app = JupyterDash(__name__)
```

```python
app.layout = html.Div([
	html.H1("Stock Price Analysis Dashboard"),
	
	html.Label("Select Stock Symbol:"),
	dcc.Dropdown(
		id='stock-selector',
		options=[
			{'label': 'AAPL', 'value': 'AAPL'},
			{'label': 'GOOGL', 'value': 'GOOGL'},
			{'label': 'MSFT', 'value': 'MSFT'},
		],
		value='AAPL'
	),
	
	dcc.Graph(id='stock-price-chart'),
])
```

```python
@app.callback(
	Output('stock-price-chart', 'figure'),
	[Input('stock-selector', 'value')]
)
def update_chart(selected_stock):
	stock_data = yf.download(selected_stock, start='2022-01-01',
							end='2023-01-01')
	fig = px.line(stock_data, x=stock_data.index, y='Close',
				title=f'{selected_stock} Stock Price Analysis')
	return fig
```

```python
if __name__ == '__main__':
	app.run_server(port=8058, mode='external')
```

```python
from dash import Dash, dcc, html, Input, Output
import pandas as pd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import numpy as np
from io import BytesIO
import base64

app = Dash(__name__)

# Tu DataFrame (df)
# Supongo que df ya está cargado con tus datos
# df = ...

# Convertir las fechas a objetos datetime de Python si es necesario
df['fecha'] = pd.to_datetime(df['fecha'])

# Encontrar los valores mínimo y máximo de FWI para todo el DataFrame
fwi_min = df['FWI'].min()
fwi_max = df['FWI'].max()

# Crear una función para el gráfico
def plot_fwi(fecha):
    fecha_seleccionada = df[df['fecha'] == fecha]

    fig, ax = plt.subplots(figsize=(10, 6), subplot_kw={'projection': ccrs.PlateCarree()})

    # Agregar los límites de los países
    ax.add_feature(cfeature.BORDERS, linestyle=':')
    ax.add_feature(cfeature.COASTLINE)
    ax.add_feature(cfeature.LAND, facecolor='white', edgecolor='black')

    # Ajustar los límites del mapa según tus necesidades
    ax.set_extent([-120, -80, 15, 35], crs=ccrs.PlateCarree())

    # Agregar los puntos sin bordes
    scatter = ax.scatter(fecha_seleccionada['longitud'], fecha_seleccionada['latitud'],
                         c=fecha_seleccionada['FWI'], cmap='coolwarm', s=100, edgecolor='none', alpha=0.7, 
                         vmin=fwi_min, vmax=fwi_max, transform=ccrs.PlateCarree())

    # Agregar barra de color general
    cbar = plt.colorbar(scatter, ax=ax, orientation='vertical', pad=0.02, aspect=50, shrink=0.9)
    cbar.set_label('FWI')

    # Formatear la fecha
    fecha_python = pd.Timestamp(fecha).to_pydatetime()
    ax.set_title(f'FWI para {fecha_python.strftime("%d-%m-%Y %H:%M")}')
    ax.set_xlabel('Longitud')
    ax.set_ylabel('Latitud')

    # Convertir la figura a una imagen en base64 para mostrarla en el navegador
    buf = BytesIO()
    plt.savefig(buf, format="png")
    plt.close(fig)
    data = base64.b64encode(buf.getbuffer()).decode("utf8")
    return f"data:image/png;base64,{data}"

# Opciones únicas de fecha para el Dropdown
fechas_unicas = df['fecha'].unique()

app.layout = html.Div([
    html.H1("Mapa interactivo de FWI"),
    dcc.Dropdown(
        id='fecha-dropdown',
        options=[{'label': str(fecha), 'value': str(fecha)} for fecha in fechas_unicas],
        value=str(fechas_unicas[0]),
    ),
    html.Img(id='fwi-map')
])

@app.callback(
    Output('fwi-map', 'src'),
    [Input('fecha-dropdown', 'value')]
)
def update_map(selected_date):
    return plot_fwi(selected_date)

if __name__ == '__main__':
    app.run_server(debug=True, port=8051)
```
